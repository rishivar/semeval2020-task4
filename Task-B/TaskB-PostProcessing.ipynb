## Post Processing for BERT and XLNet

```python
import pandas as pd

filename = #['bert', 'xlnet']

#Initialise file name from any of the above list elements to run post processing for the particular output file.

df = pd.read_csv(filename +'./csv', sep = '\t')
df # 0 - Class 0 score, 1 - Class 1 score
```

```python
def func(r):
    if(r['1']>r['0']):
        return [r['1'], 1]
    else:
        return [r['0'], 0]

df['tar'] = df.apply(lambda r: func(r), 1)
df = df.drop(['1','0'], 1)

l = df['tar']

res = []

def func(x, y, z):
    mark = list()
    if(x[1] == 1):
        mark.append(x)
    if(y[1] == 1):
        mark.append(y)
    if(z[1] == 1):
        mark.append(z)
    elif(x[1] == y[1] == z[1] == 0):
        mark.append(x)
        mark.append(y)
        mark.append(z)
        return min(mark)
    return max(mark)

for i in range(0, len(l), 3):
    l[i].append('A')
    l[i+1].append('B')
    l[i+2].append('C')
    res.append(func(l[i], l[i+1], l[i+2]))
    
res
```

```python
final = pd.DataFrame(res, columns = [filename+'a', '2', filename+'b'])
del final['2']
final.to_csv('Model ouputs/'+ filename + 'pp.csv', header = False, index = False)
```
